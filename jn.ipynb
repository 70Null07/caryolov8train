```python
from PIL import Image, ImageDraw
from torchvision import transforms

image_path = 'img\output_0028.png'
image = Image.open(image_path)

transform = transforms.Compose([
    transforms.Resize((640, 640))
])

image = transform(image)

# Run inference on the image
results = model(image)  # Get predictions

# Draw a point in the center of each detected car
draw = ImageDraw.Draw(image)
for box in results[0].boxes.xyxy:
    x, y, w, h = box.tolist()  # Get coordinates of the detected object
    center_x = int(x + (w-x)/2)
    center_y = int(y + (h-y)/2)
    #draw.rectangle([x, y, w, h], outline ="red")
    #draw.point((center_x, center_y), fill='red')  # Draw a red point at the center
    draw.ellipse((center_x-7, center_y-7, center_x+7, center_y+7), fill='black')  # Draw a black circle with a radius of 7 pixels
    draw.ellipse((center_x-5, center_y-5, center_x+5, center_y+5), fill='yellow')  # Draw a yellow circle with a radius of 5 pixels inside the black one

display(image)

# Save the modified image
image.save('result\\result_0028.png')
```
